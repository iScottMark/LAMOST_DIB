# Generate the bash command according to the number of the blade servers

```python
# num_server = 6
# num_file = 2_188_240
num_server = 2
num_file = 688_491
interval = num_file // num_server
command = './shell_scripts/multiple_tasks.sh'
shell_cmd = []

for i in range(num_server):
    if not i == num_server - 1:
        shell_cmd.append(f'{command} {i * interval} {(i + 1) * interval} 42')
    else:
        shell_cmd.append(f'{command} {i * interval} {num_file} 42')

# print(f'{shell_cmd[0]}')
# print(f'ssh maxiaoxiao@172.17.0.121 \"{shell_cmd[1]}\"')
# print(f'ssh maxiaoxiao@172.17.0.122 \"{shell_cmd[2]}\"')
# print(f'ssh maxiaoxiao@172.17.0.123 \"{shell_cmd[3]}\"')
# print(f'ssh maxiaoxiao@172.17.0.124 \"{shell_cmd[4]}\"')
# print(f'ssh maxiaoxiao@172.17.0.126 \"{shell_cmd[5]}\"')
# print(f'ssh maxiaoxiao@172.17.0.127 \"{shell_cmd[6]}\"')
```
